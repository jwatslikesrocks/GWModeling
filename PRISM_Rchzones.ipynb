```python
#Specify recharged coefficients for each recharge zone
zone1=[0.01, 0.03]
zone2=[0.005, 0.01]
zone3=[0.01, 0.025]
zone4=[0.01, 0.2]
zone5=[0.001, 0.03]
zone6=[0.01, 0.15]
```

```python
flow=pd.read_csv('Blancoflow.csv')
#import Blanco river flow (monthly average) in units of ft3/s; CSV must have 3 columns: Date/SP/BLAflow
```

```python
ppt=pd.read_csv('PRISM_ft-d.csv')
pptx=ppt.set_index('GridID')
pptx
#import PRISM precipitation data (monthly) in units of ft/d; creat df "pptx" with GridID specified as index; 1st CSV column must have prism GRIDID; than each subsequent column recharge values for each stress period
```

```python
nodes=pd.read_csv('Rechargedomain_spatialjoin.csv')
#import table containing entries for every recharge node within the recharge domain and fields for recharge zone and PRISM gridID
#TODO: generalize table format to allow python to pull data based on column name regardless of location
```

```python
nodes['GridID']=nodes['GridID'].apply(np.int64)
nodes.set_index('GridID')
nodes
```

```python
a=len(flow)
b=a-1
#defines array with number of stress periods from blanco river flow array
```

```python
d=np.arange(1,a+1)
#create increasing integer array with entry for each stress period
```

```python
c=len(nodes.node)
#defines array with number of nodes in recharge domain
```

```python
x=nodes.set_index('GridID')
df1=x
df2=pptx
df3=pd.merge(df1,df2,left_index=True,right_index=True)
#Merge ppt and nodes dataset into new dataset called 'df3'
#for df3.iloc[:,7] now need to apply conditions to created 
    #if flow.BLAflow[0] > 40
```

```python
#for loop apply recharge coefficients to recharge dataframe; 
for i in flow.index:
    for j in range(len(df3)):
        if df3.Rchzone.iloc[j] == 2:   
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone2[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone2[0]
        if df3.Rchzone.iloc[j] == 1:
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone1[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone1[0]
        if df3.Rchzone.iloc[j] == 3:
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone3[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone3[0]
        if df3.Rchzone.iloc[j] == 5:
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone5[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone5[0]     
        if df3.Rchzone.iloc[j] == 6:
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone6[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone6[0]  
        if df3.Rchzone.iloc[j] == 4:
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone4[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*zone4[0] 
```
