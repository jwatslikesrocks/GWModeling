```python
#Specify resired recharged coefficients for each recharge zone
Bladwn=[0.01, 0.15]
Blaup=[0.01, 0.025]
Onion=[0.01, 0.05]
Downgrad=[0.01, 0.025]
```

```python
x=nodes.set_index('GridID')
df1=x
df2=pptx
df3=pd.merge(df1,df2,left_index=True,right_index=True)
#Merge ppt and nodes dataset into new dataset called 'df3'
#for df3.iloc[:,7] now need to apply conditions to created 
    #if flow.BLAflow[0] > 40
```

```python
#for loop apply recharge coefficients to recharge dataframe; to do: turn recharge coefficients into separate input variables at beginning of script
for i in flow.index:
    for j in range(len(df3)):
        if df3.Rchzone.iloc[j] == 2:   
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*Blaup[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*Blaup[0]
        if df3.Rchzone.iloc[j] == 1:
            if flow.BLAflow[i] > 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*Bladwn[1]
            elif flow.BLAflow[i] < 40:
                df3.iloc[j,i+7]=df3.iloc[j,i+7]*Bladwn[0]
        if df3.Rchzone.iloc[j] == 3:
                    if flow.BLAflow[i] > 40:
                        df3.iloc[j,i+7]=df3.iloc[j,i+7]*Onion[1]
                    elif flow.BLAflow[i] < 40:
                        df3.iloc[j,i+7]=df3.iloc[j,i+7]*Onion[0]
        if df3.Rchzone.iloc[j] == 5:
                    if flow.BLAflow[i] > 40:
                        df3.iloc[j,i+7]=df3.iloc[j,i+7]*Downgrad[1]
                    elif flow.BLAflow[i] < 40:
                        df3.iloc[j,i+7]=df3.iloc[j,i+7]*Downgrad[0]        
```
